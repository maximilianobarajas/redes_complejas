```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()

with open("red.txt",'r') as file:
    for line in file:
        nodes = line.strip().split()
        G.add_edge(nodes[0],nodes[1])
print(G)
print(G.nodes)
print(G.edges)

nx.draw(G, with_labels=True, node_color='blue', node_size=100, font_size=12, font_weight='bold')
plt.show()
```

```python
import csv
G= nx.Graph()
with open("red2.csv",'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if len(row) != 0:
            G.add_edge(row[0],row[1])
print(G)
print(G.nodes)
print(G.edges)
nx.draw(G, with_labels=True, node_color='red', node_size=100, font_size=12, font_weight='bold')
plt.show()
```
